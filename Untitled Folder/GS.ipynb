```python

df = pd.read_excel('utility/dataa.xls')
# Define the order of the cities based on the GDP rankings
city_order = ['Guangdong', 'Jiangsu', 'Shandong', 'Zhejiang', 'Henan', 'Sichuan', 'Hubei', 'Hunan', 'Anhui', 'Fujian', 'Shanghai', 'Beijing', 'Liaoning', 'Shaanxi', 'Hebei', 'Tianjin', 'Chongqing', 'Guangxi', 'Yunnan', 'Jilin', 'Heilongjiang', 'Guizhou', 'Ningxia', 'Hainan', 'Shanxi', 'Gansu', 'Qinghai', 'Inner Mongolia', 'Jiangxi', 'Xinjiang']

# Reorder the cities in the DataFrame based on the GDP rankings
df = df.set_index('Year')
df = df.reindex(city_order)

# Create bar graphs for each wage column
plt.figure(figsize=(12, 6))

# Total (4-15)
plt.subplot(1, 2, 1)
plt.bar(df.index, df['Total (4-15)'])
plt.xticks(rotation=90)
plt.ylabel('Average Wage (yuan)')
plt.title('Avg Wage in Urban Non-Private Units')

# Manufacturing (4-15)
plt.subplot(1, 2, 2)
plt.bar(df.index, df['Manufacturing (4-15)'])
plt.xticks(rotation=90)
plt.ylabel('Average Wage (yuan)')
plt.title('Avg Wage in Urban Non-Private Units in Manufacturing')


plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

# Total (4-16)
plt.subplot(1, 2, 1)
plt.bar(df.index, df['Total (4-16)'])
plt.xticks(rotation=90)
plt.ylabel('Average Wage (yuan)')
plt.title('Avg Wage in Urban Private Units')

# Manufacturing (4-16)
plt.subplot(1, 2, 2)
plt.bar(df.index, df['Manufacturing (4-16)'])
plt.xticks(rotation=90)
plt.ylabel('Average Wage (yuan)')
plt.title('Avg Wage in Urban Private Units in Manufacturing')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_excel('utility/dataa.xls')

# Define the coastal and inland city groups
coastal_cities = ['Beijing', 'Tianjin', 'Hebei', 'Liaoning', 'Shanghai', 'Jiangsu', 'Zhejiang', 'Fujian', 'Shandong', 'Guangdong', 'Guangxi', 'Hainan']
inland_cities = ['Shanxi', 'Inner Mongolia', 'Jilin', 'Heilongjiang', 'Anhui', 'Jiangxi', 'Henan', 'Hubei', 'Hunan', 'Chongqing', 'Sichuan', 'Guizhou', 'Yunnan', 'Shaanxi', 'Gansu', 'Qinghai', 'Ningxia', 'Xinjiang']

# Compute the average wage for each column for the coastal and inland city groups
coastal_avg_total_15 = df[df['Year'].isin(coastal_cities)]['Total (4-15)'].mean()
inland_avg_total_15 = df[df['Year'].isin(inland_cities)]['Total (4-15)'].mean()

coastal_avg_total_16 = df[df['Year'].isin(coastal_cities)]['Total (4-16)'].mean()
inland_avg_total_16 = df[df['Year'].isin(inland_cities)]['Total (4-16)'].mean()

coastal_avg_manuf_15 = df[df['Year'].isin(coastal_cities)]['Manufacturing (4-15)'].mean()
inland_avg_manuf_15 = df[df['Year'].isin(inland_cities)]['Manufacturing (4-15)'].mean()

coastal_avg_manuf_16 = df[df['Year'].isin(coastal_cities)]['Manufacturing (4-16)'].mean()
inland_avg_manuf_16 = df[df['Year'].isin(inland_cities)]['Manufacturing (4-16)'].mean()

# Add Vietnam and India to the data
vietnam_manuf = 21224.51
india_manuf = 23581.61

# Create a bar graph comparing the average wage for the Total (4-15) column
plt.bar(['Coastal', 'Inland'], [coastal_avg_total_15, inland_avg_total_15])
plt.title('Average Wage (Total, 4-15)')
plt.ylabel('Average Wage (yuan)')
plt.show()

# Create a bar graph comparing the average wage for the Total (4-16) column
plt.bar(['Coastal', 'Inland'], [coastal_avg_total_16, inland_avg_total_16])
plt.title('Average Wage (Total, 4-16)')
plt.ylabel('Average Wage (yuan)')
plt.show()


# Create a bar graph comparing the average wage for the Manufacturing (4-15) column
plt.bar(['China (Coastal)', 'China (Inland)', 'Vietnam', 'India'], [coastal_avg_manuf_15, inland_avg_manuf_15, vietnam_manuf, india_manuf])
plt.title('Avg Wage in Urban Non-Private Units in Manufacturing vs. Vietnam & India')
plt.ylabel('Average Wage (yuan)')
plt.show()

# Create a bar graph comparing the average wage for the Manufacturing (4-16) column
plt.bar(['China (Coastal)', 'China (Inland)', 'Vietnam', 'India'], [coastal_avg_manuf_16, inland_avg_manuf_16, vietnam_manuf, india_manuf])
plt.title('Avg Wage in Urban Private Units in Manufacturing vs. Vietnam & India')
plt.ylabel('Average Wage (yuan)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read the data from the CSV file
data = pd.read_excel('utility/data.xls')
# drop the rows with missing values
data = data.dropna()

# create line plot
plt.figure(figsize=(12, 6))
plt.plot(data["Year"], data["Total (4-15)"], color="maroon")
plt.plot(data["Year"], data["Total (4-16)"], color="black")
plt.xlabel("Year")
plt.ylabel("Average Wage (Yuan)")
plt.title("Comparison of Average Wages by Sector")
plt.legend(["Urban Non-Private Units", "Urban Private Units"])
plt.show()

# calculate percentage change
data["Percentage Change (4-15)"] = data["Total (4-15)"].pct_change() * 100
data["Percentage Change (4-16)"] = data["Total (4-16)"].pct_change() * 100

# create line plot for China different sectors and different nations
# Data for Mexico and OECD-Total
mexico_percent_change = [-5.85, 0.90, -1.80, 0.07, 0.19, 1.30, -1.17, -0.26, 2.23, 0.95]
oecd_percent_change=[0.16, 0.28, 0.03, 0.28, 0.74, 1.77, 1.71, 0.87, 0.93, 1.73]


plt.figure(figsize=(12, 6))
plt.plot(data["Year"], data["Percentage Change (4-15)"], color="maroon")
plt.plot(data["Year"], data["Percentage Change (4-16)"], color="black")
plt.plot(range(2009, 2019), [-1.51, 2.36, 3.13, 3.12, 1.28, 3.55, 3.48, 1.13, 3.45, 3.62], color="grey")
plt.plot(data["Year"], mexico_percent_change, color="blue")
plt.plot(data["Year"], oecd_percent_change, color="green")
plt.xlabel("Year")
plt.ylabel("Percentage Change (%)")
plt.title("Growth of Wages of Different Sectors in Percentages")
plt.ylim(-10, None)
plt.legend(["Urban Non-Private Units (China)", "Urban Private Units (China)", "United States", "Mexico", "OECD-Total"])
plt.show()

# create box plot
plt.figure(figsize=(12, 6))
plt.boxplot([data["Total (4-15)"], data["Total (4-16)"]], labels=["Urban Non-Private Units", "Urban Private Units"])
plt.xlabel("Sector")
plt.ylabel("Average Wage (Yuan)")
plt.title("Box Plot of Average Wages by Sector")
plt.show()

# create scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(data["Year"], data["Total (4-15)"], color="maroon")
plt.scatter(data["Year"], data["Total (4-16)"], color="black")
plt.xlabel("Year")
plt.ylabel("Average Wage (Yuan)")
plt.title("Comparison of Average Wages by Sector")
plt.legend(["Urban Non-Private Units", "Urban Private Units"])
plt.show()

# create bar chart
plt.figure(figsize=(12, 6))
plt.bar(data["Year"], data["Total (4-15)"], color="maroon", alpha=0.5)
plt.bar(data["Year"], data["Total (4-16)"], color="black", alpha=0.5)
plt.xlabel("Year")
plt.ylabel("Average Wage (Yuan)")
plt.title("Comparison of Average Wages by Sector")
plt.legend(["Urban Non-Private Units", "Urban Private Units"])
plt.show()
```
